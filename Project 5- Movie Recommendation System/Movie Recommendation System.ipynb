# Building Recommender System

```python
c= df['vote_average'].mean()
c
```

```python
m= df['vote_count'].quantile(0.9)   #90% of total no. of votes
m
```

```python
# ...Criteria...

qm= df[(df['runtime']>150) & (df['revenue']>300000000) & (df['homepage'].notna())]

qm.head()
```

```python
qm.shape
```

```python
qm['vote_count']>= m
```

```python
qm= qm[qm['vote_count']>= m]

qm.shape
```

# Now we have to select Top 10 Movies

IMDB Rating calculated

m= quantile

c= mean (avg)

```python
def wr(x,m=m,c=c):
    v = x['vote_count']
    r = x['vote_average']
    return (v/(v+m)*r) + (m/(m+v)*c)
```

```python
qm['score']= qm.apply(wr, axis=1)
```

```python
qm= qm.sort_values('score', ascending=False)
```
