```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
```

```python
df = pd.read_csv('general_data.csv')
```

```python
df
```

```python
df['Attrition']=np.where(df['Attrition']=="True",1,0)
```

```python
df
```

```python

l=['Age', 'Attrition',   'DistanceFromHome',
        'EmployeeCount',
       'JobLevel',  'MonthlyIncome','Education',
       'NumCompaniesWorked',  'PercentSalaryHike', 'StandardHours',
       'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
```

```python
k=['BusinessTravel', 'Department','EducationField','Gender', 'MaritalStatus']
```

```python
for i in k:
    af=pd.get_dummies(df[i])
    df=pd.concat([df,af],axis=1)
    df.drop(i,axis=1,inplace=True)
```

```python
df
```

```python
df['NumCompaniesWorked']=np.where(df['NumCompaniesWorked'].isnull(),df['NumCompaniesWorked'].median(),df['NumCompaniesWorked'])
df['TotalWorkingYears']=np.where(df['TotalWorkingYears'].isnull(),df['TotalWorkingYears'].median(),df['TotalWorkingYears'])
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.corr(),annot =True)
```

```python
plt.scatter(df.PercentSalaryHike,df.TrainingTimesLastYear)
```

```python
r=[]
for i in l:
    print("Correlation between ",i, "and all other various features:")
    for j in l:
        if j not in r :
            x , p = st.pearsonr(df[i],df[j])
            print(i," vs ", j)
            print("P value is", p)
            print("Correlation between ",i, "and ",j, "is ",x)
            
            if p < 0.05:
                print("P-value is less than 0.05 which results in not a viable correlation")
            else:
                print("P-value is greater than 0.05 which results in viable correlation")
                
    r.append(i)
    print("")
```
